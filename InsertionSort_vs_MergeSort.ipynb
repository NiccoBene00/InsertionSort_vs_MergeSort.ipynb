##Cenni Teorici

### Insertion Sort

L'algoritmo **Insertion-Sort** è un semplice algoritmo di ordinamento "sul posto" basato sull'idea di costruire progressivamente una sequenza ordinata spostando gli elementi uno alla volta nella posizione corretta. È simile al modo in cui gli esseri umani ordinano le carte in mano durante un gioco.


1. L'algoritmo considera l'array diviso in due parti: una parte iniziale ordinata e una parte finale disordinata.
2. In ogni passo, il primo elemento della parte disordinata viene preso e inserito nella posizione corretta della parte ordinata.
3. Questo processo si ripete fino a quando tutti gli elementi non sono stati inseriti nella parte ordinata.


Pseudocodice di Insertion Sort:



```
INSERTION-SORT(A)
    for j = 2 to A.length
        key = A[j]
        i = j - 1
        while i > 0 and A[i] > key
            A[i + 1] = A[i]
            i = i - 1
        end while
        A[i + 1] = key
```

#### Complessità temporale

##### Caso migliore
- **Descrizione**: Il caso migliore si verifica quando l'array è già ordinato.
- **Complessità**: In questo caso, il ciclo interno (`while`) non effettua alcun confronto o spostamento.
- **Ordine di crescita**: $O(n)$.

##### Caso medio
- **Descrizione**: In media, ogni elemento deve essere confrontato con circa metà degli elementi nella parte già ordinata.
- **Complessità**: Ogni inserimento richiede un numero crescente di confronti man mano che ci si sposta nell'array.
- **Ordine di crescita**: $O(n²)$.

##### Caso peggiore
- **Descrizione**: Il caso peggiore si verifica quando l'array è ordinato in ordine decrescente.
- **Complessità**: Ogni elemento deve essere confrontato e spostato fino alla posizione iniziale.
- **Ordine di crescita**: $O(n²)$.

#### Complessità spaziale
- **Tipo di algoritmo**: In-place.
- **Descrizione**: L'algoritmo non richiede spazio aggiuntivo significativo oltre alla memoria necessaria per il vettore di input.
- **Ordine di crescita**: $O(1)$.

### Merge Sort

**Merge-Sort** è un algoritmo di ordinamento basato sul paradigma di **divide et impera**. È particolarmente efficiente per ordinare grandi insiemi di dati grazie alla sua complessità temporale ottimale nei casi migliori, medi e peggiori. Il funzionamento logico dell'algoritmo è il seguente:


1. **Divide**: Il vettore da ordinare viene diviso a metà ricorsivamente fino a ottenere sottosequenze di un solo elemento (che sono appunto ordinate).
2. **Conquista**: Le sottosequenze vengono combinate ordinandole in modo incrementale fino a ricostruire l'intero array ordinato.
3. **Combina**: Le due sottosequenze ordinate vengono fuse (merge) in un'unica sequenza ordinata.

Pseudocodice di Merge Sort:

```
MERGE-SORT(A, p, r)
    if p < r
        q = ⌊(p + r) / 2⌋
        MERGE-SORT(A, p, q)
        MERGE-SORT(A, q + 1, r)
        MERGE(A, p, q, r)

MERGE(A, p, q, r)
    n1 = q - p + 1
    n2 = r - q
    crea due nuovi array L[1...n1 + 1] e R[1...n2 + 2]
    for i = 1 to n1
        L[i] = A[p + i - 1]
    for j = 1 to n2
        R[j] = A[q + j]
    L[n1 + 1] = ∞
    R[n2 + 1] = ∞
    i = 1
    j = 1
    for k = p to r
        if L[i] <= R[j]
            A[k] = L[i]
            i = i + 1
        else
            A[k] = R[j]
            j = j + 1
```

#### Complessità temporale

##### Caso migliore
- **Descrizione**: Merge Sort divide sempre l'array a metà e combina le sottosequenze in modo ordinato. Il numero di operazioni è lo stesso indipendentemente dall'ordine iniziale dei dati.
- **Complessità**: $O(n log n)$, dove:
  - $( log n )$: Rappresenta il numero di livelli della ricorsione, poiché l'array viene diviso in due parti a ogni passo.
  - $( n )$: Corrisponde al costo per fondere gli elementi in ogni livello.

##### Caso medio
- **Descrizione**: Anche nel caso medio, l'algoritmo divide e combina le sottosequenze nello stesso modo indipendentemente dall'ordine iniziale.
- **Complessità**: $O(n log n)$. Non sussistono quindi variazioni significative rispetto al caso migliore.

##### Caso peggiore
- **Descrizione**: Il caso peggiore si verifica nello stesso modo in cui si verifica il caso migliore, poiché l'algoritmo non dipende dall'ordine iniziale dell'array.
- **Complessità**: $O(n log n)$.



#### Complessità spaziale

- **Descrizione**: Merge Sort non è un algoritmo **in-place**, poiché richiede spazio aggiuntivo per memorizzare i sottovettori durante il processo di fusione.
- **Spazio richiesto**:
  - $O(n)$: Per creare i sottovettori necessari alla fusione.
  - L'algoritmo utilizza ulteriore spazio per gestire le chiamate ricorsive. Questo contribuisce con un fattore di $O(log n)$ allo stack delle chiamate ricorsive.

####Classe `SortAlgorithm`

Segue una breve descrizione dei metodi implementati:
- `def __init__(self)`: costruttore della classe, inizializza il vettore su cui viene eseguito l'ordinamento come una lista vuota;

- `def insertionSort(self, input)`: prende in ingresso il parametro `input` come vettore da ordinare, dunque applica la logica di ordinamento di insertion sort;

- `def mergeSort(self, input)`: prende in ingresso il parametro `input` come vettore da ordinare, dunque definisce al suo interno un metodo ricorsivo `merge_sort_rec(arr)` che implementa la logica di merge sort;

- `def merge_sort_rec(self, arr)`: applica un'implementazione di Merge-Sort senza l'utilizzo dei valori sentinella inizializzati a infinito per gestire gli elementi rimanenti nei sottovettori una volta che uno dei due si esaurisce. La presenza degli ultimi due cicli `while` che differiscono dallo pseudocodice si deve proprio a questo nuovo approccio;

**Osservazione:** gli ordinamenti a cura degli algoritmi presentati vengono eseguiti su liste (vettori) di interi non negativi per facilitare gli esperimenti condotti.

####Classe `SortingComparison`

Segue una breve descrizione dei metodi implementati:
- `def __init__(self)`: inizializza come attributo di classe un'istanza della classe SortAlgorithm.

- `def generateInput(self, size, structure)`: genera un vettore di input che abbia una dimensione specificata dal parametro `size` e una certa proprietà strutturale `structure`. Nello specifico con proprietà strutturale si fa riferimento all'ordine con il quale il vettore input generato si presenterà:
  - Randomico
  - Ordinato
  - Poco ordinato
  - Ordinato al contrario

- `def compareAlgorithms(self, sizes, structures)`: itera su ogni possibile dimensione e struttura richiesta, dunque ordina ogni input associato a tali caratteristiche con Insertion-Sort e Merge-Sort effettuanto le chiamate ai metodi necessari. Genera un vettore stringa risulato in cui si riportano le caratteristiche dell'input iniziale, e le tempistiche dei due algoritmi a confronto.

````python

class SortAlgorithm:

  def __init__(self):
    print("Inizializing sort algorithms class...")
    self.vector = []

  def getResult(self):
    return self.vector

  def resetResult(self):
    self.vector = []

  def insertionSort(self, input):
    self.vector = input
    for j in range(1, len(self.vector)):
      key = self.vector[j]
      i = j - 1
      while i >= 0 and self.vector[i] > key:
        self.vector[i + 1] = self.vector[i]
        i = i - 1
      self.vector[i + 1] = key



  def mergeSort(self, input):
    #copy parameter "input" into self intern vector through an indipendent copy
    self.vector = input[:]

    #recursive method for sorting
    def merge_sort_rec(arr):
      if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid] #create an indipendent list for the first half of
                              #"arr"'s element
        right_half = arr[mid:] #create an indipendent list for the second half
                               #of "arr"'s element

        #recursion on two halfes
        merge_sort_rec(left_half)
        merge_sort_rec(right_half)

        #merge two sorted half
        i = j = k = 0 # i := scanner of "left_half"
                      # j := scanner of "right_half"
                      # k := scanner of "arr"
        while i < len(left_half) and j < len(right_half): #until there are
                                                          #elements in both halfes


          if left_half[i] < right_half[j]:
            arr[k] = left_half[i]
            i += 1
          else:
            arr[k] = right_half[j]
            j += 1
          k += 1

        #copy remain elements into the result
        while i < len(left_half):
          arr[k] = left_half[i]
          i += 1
          k += 1

        while j < len(right_half):
           arr[k] = right_half[j]
           j += 1
           k += 1

    #start sorting on self intern vector
    merge_sort_rec(self.vector)



````

````python
import time
import random

class SortingComparison:
    def __init__(self):
        self.algorithms = SortAlgorithm()

    def generateInput(self, size, structure):
        if structure == "random":
            return [random.randint(1, 10000) for _ in range(size)]

        elif structure == "almost_sorted":
            sorted_arr = sorted([random.randint(1, 10000) for _ in range(size)])
            for _ in range(max(1, size // 10)): # introduce varibale number of
                                                # swaps
                i, j = random.sample(range(size), 2) #assign two random distinct
                                                     #index to the variable i,j
                 #execute the swap on "arr" list at index i,j
                sorted_arr[i], sorted_arr[j] = sorted_arr[j], sorted_arr[i]
            return sorted_arr

        elif structure == "reversed":
            return sorted([random.randint(1, 10000) for _ in range(size)],
                          reverse=True)

        elif structure == "sorted_list":
            return sorted([random.randint(1, 10000) for _ in range(size)])

        else:
            raise ValueError("Invalid structure type!")

    def compareAlgorithms(self, sizes, structures):
        results = []

        for structure in structures:
            for size in sizes:
                input_data = self.generateInput(size, structure)

                # Test Insertion Sort
                start_time = time.perf_counter()
                self.algorithms.insertionSort(input_data)
                insertion_time = time.perf_counter() - start_time

                # Test Merge Sort
                start_time = time.perf_counter()
                self.algorithms.mergeSort(input_data)
                merge_time = time.perf_counter() - start_time

                # append results
                results.append({
                    "size": size,
                    "structure": structure,
                    "insertion_sort_time": insertion_time,
                    "merge_sort_time": merge_time
                })
        return results

    def analyzeResults(self, results):
        print("\n--- Sorting Algorithm Comparison ---\n")
        for result in results:
            size = result["size"]
            structure = result["structure"]
            insertion_time = result["insertion_sort_time"]
            merge_time = result["merge_sort_time"]

            print("***********************************************")
            print(f"[Input Size: {size} - Structure: {structure}]")
            print(f"Insertion Sort Time: {insertion_time:.6f} seconds")
            print(f"Merge Sort Time: {merge_time:.6f} seconds")

            if insertion_time < merge_time:
                print("Faster: Insertion Sort")
            else:
                print("Faster: Merge Sort")



````

##Esperimenti e dati utilizzati
La comparazione tra i due algoritmi di ordinamento Insertion-Sort e Merge-Sort viene eseguita attraverso l'analisi del comportamento di questi quando vengono presi in considerazioni input di `size` progressiva e `structure` diversa. Nel dettaglio del codice che segue si prendono come riferimento cardinalità degli input pari a 50, 100, 500, 1000, 5000 elementi (interi non negativi nell'intervallo [1-10000]) con struttura randomizzata/ordinata/poco ordinata/ordinata al contrario.  
L'output che segue fornisce una panoramica circa l'esecuzione temporale dei due diversi ordinamenti.


````python
import random
import builtins

#with the following lines of code we avoid to overwrite built-in python function
#"sorted"; so doing we don't occur in "sorted is no longer callable" errror
if 'sorted' in globals() and not callable(globals()['sorted']):
    sorted = builtins.sorted

comparison = SortingComparison()
sizes = [50, 100, 500, 1000, 5000]
structures = ["random", "almost_sorted", "reversed", "sorted_list"]

results = comparison.compareAlgorithms(sizes, structures)
comparison.analyzeResults(results)
````

##Analisi dei risultati sperimentali

1.  Struttura randomizzata: si osserva che l'algoritmo Merge-Sort è sistematicamente più veloce temporalmente e guadagna un vantaggio crescente su input di dimensione maggiore.
    * per size=50 ad esempio la differenza temporale tra i due algoritmi è minima (circa 0.00005s) ma questa aumenta drasticamente per size=5000 (Merge-Sort è 100 volte più veloce);
    * si nota quindi che, in linea con la complessità teorica, Insertion-Sort peggiora rapidamente come $O(n²)$, mentre Merge-Sort come $O(n log n)$;


2.  Struttura poco ordinata
    * per input di size piccola (inferiore a 100 elementi) Insertion-Sort sfrutta una complessità $O(n)$, quindi risulta più veloce;
    * per input di size maggiore (sopra i 500 elementi) Merge-Sort diviene dominante grazie alla sua complessità $O(nlogn)$;


3.  Struttura ordinata al contrario: Merge-Sort è sempre più veloce, a discapito della size dell'input presentato.
    * Insertion-Sort soffre quindi del caso peggiore $O(n²)$ causa l'inversione totale dell'input;
    * Merge-Sort mantiene invece tempi ridotti grazie alla sua struttura ricorsiva;


4.  Struttura ordinata: Insertion-Sort supera Merg-Sort in ogni caso, confermando la complessità teorica nel suo caso migliore $O(n)$. La riduzione drastica delle tempistiche è dovuta al numero al limitatissimo numero di confronti tra elementi dell'input; proprietà di cui non gode Merge-Sort.
